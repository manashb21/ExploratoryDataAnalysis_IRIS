<h2> 2-D scatter Plot</h2>

```python
#2-D scatter plot 
iris.plot(kind = 'scatter', x = 'sepal.length', y = 'sepal.width')
plt.show()
```

The plot above does not really relay much information other than the range of the lengths and width of sepals.
So, we can color these points based on the species.

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue = 'species', height = 5) \
    .map(plt.scatter, "sepal.length", "sepal.width")\
    .add_legend()
plt.show()
```

<b>Observations</b>
<ol>
    <li>The blue points or Setosa are well separated from other two.
    <li>Separating Versicolor and Virginica is harder as they have considerable overlap.

 Using 3D plot is not so easy to view on 2D planes. So we can use Pair Plots to visualize all the datas.

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris, hue = "species", height = 3) #dataset, colored by which label, height of plot
plt.show()
```

Focusing on top 6 plots above the diagonal, 

<b>Observations</b>
<ul>
    <li>Petal lengths and petal widths are able to separate flowers much better and sepal lenghts and sepal widths.
    <li>While Setosa can be easily identified (linearly separable), Virginica and Versicolor have some overlap (almost lineraly separable)
    <li>We can find "lines" and "if-else" conditions to build a simple model to classify the flower types.
</ul>

<b>Limitations of Pair Plots</b>

For a dataset with 4 features, we have 4C2 plots. For dataset with n features, we would have nC2 plots. 
That is not efficient. 

# Histogram, PDF and CDF

```python
import numpy as np
iris_setosa = iris.loc[iris['species'] == "Setosa"]
iris_virginica = iris.loc[iris['species'] == "Virginica"]
iris_versicolor = iris.loc[iris['species'] == "Versicolor"]

plt.plot(iris_setosa["petal.length"], np.zeros_like(iris_setosa['petal.length']))
plt.plot(iris_versicolor["petal.length"], np.zeros_like(iris_versicolor['petal.length']))
plt.plot(iris_virginica["petal.length"], np.zeros_like(iris_virginica['petal.length']))

plt.show()
```

```python
#creating a Histogram of Petal lengths of flowers 
sns.FacetGrid(iris, hue = "species", height = 5)\
    .map(sns.histplot, "petal.length")\
    .add_legend()
    
plt.show()
    
```

X axis is petal length, Y axis is the count of how many petal lengths in that range. 

The smooth curve that can be made on the graphs is called PDF (Probability distribution function).

**PDF** - A smooth Histogram, a denisty plot. 

```python
sns.FacetGrid(iris, hue = "species", height = 5)\
    .map(sns.distplot, "petal.length")\
    .add_legend()
    
plt.show()
    
```

The smooth line above is the PDF, or it can also be called a desnity plot. 

1. No Setosa flowers beyond petal length of 2, so simple if condition can be used for it. 
2. PDF of Versicolor and Virginica are overlapping, when simple if logic is used, it won't provide accurate result. 



```python
sns.FacetGrid(iris, hue = "species", height = 5)\
    .map(sns.distplot, "petal.width")\
    .add_legend()
    
plt.show()
```

```python
sns.FacetGrid(iris, hue = "species", height = 5)\
    .map(sns.distplot, "sepal.length")\
    .add_legend()
    
plt.show()
```

```python
sns.FacetGrid(iris, hue = "species", height = 5)\
    .map(sns.distplot, "sepal.width")\
    .add_legend()
    
plt.show()
```

petal length > petal width >> sepal length >> sepal width 

Looking at the PDFs, the two features, petal length and petal width are more useful than others in distinguishing flowers.

# CDF (Cumulative Distribution Function)

```python
counts, bin_edges = np.histogram(iris_setosa['petal.length'], bins = 10, density = True)

pdf = counts/sum(counts)
print(pdf)
print(bin_edges)

#compute cdf
cdf = np.cumsum(pdf) #cumulative sum
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
```

```python
#setosa
counts, bin_edges = np.histogram(iris_setosa['petal.length'], bins = 10, density = True)

pdf = counts/sum(counts)
print(pdf)
print(bin_edges)

#compute cdf
cdf = np.cumsum(pdf) #cumulative sum
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

#versicolor
counts, bin_edges = np.histogram(iris_versicolor['petal.length'], bins = 10, density = True)

pdf = counts/sum(counts)
print(pdf)
print(bin_edges)

#compute cdf
cdf = np.cumsum(pdf) #cumulative sum
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

#virginica
counts, bin_edges = np.histogram(iris_virginica['petal.length'], bins = 10, density = True)

pdf = counts/sum(counts)
print(pdf)
print(bin_edges)

#compute cdf
cdf = np.cumsum(pdf) #cumulative sum
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
```
